## Objective

We are deploying a web based product for informing the public as well as elected officials towards the San Diego County 3-4-50 initiative.  In short the message is very simple: 

* __3:__ poor diet, lack of physical activity, and tobacco use _are lifestyle factors that contribute to_
* __4:__ cancer, heart disease, stroke, and type 3 diabetes _which result in_
* __50:__ over 50% of deaths in San Diego County  

The implication of these simple observations are astounding.  These chronic diseases cost the San Diego tax-payer over __4 billion dollars__ in expenditures in 2007 alone.  

There are things we can do about this, all of which require education of the population and most of them require money.  For this reason we have started a hybrid old school/new school letter/email writing campain to raise awareness amongst taxpayers and their representitives.

```python
import pandas as pd

def prettify_ax(ax):
    ax.grid(b=False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
colors = plt.rcParams['axes.color_cycle']
plt.rcParams['font.size'] = 13
```

## Calculating lives saved

Read and clean up the data

```python
f = './data/death-demographics.csv'
deaths = pd.read_csv(f, index_col=[0,1,2])
deaths.index = deaths.index.droplevel(0)
deaths = deaths.rename(index={'Elliot-\xc2\x90Navajo': 'Elliot-Navajo'})
deaths['rate_all'] = (deaths['all'] / deaths['est_pop']) * 100000
```

Calculate the percentage of 3-4-50 deaths

```python
sd_rate_3450 = (deaths['deaths3450'].unstack().sum().astype(float) / 
                deaths['est_pop'].unstack().sum()) * 100000
sd_rate_all = (deaths['all'].unstack().sum().astype(float) / 
               deaths['est_pop'].unstack().sum()) * 100000

pct_chronic = ((sd_rate_3450) / sd_rate_all).ix['2000':'2010']
```

Calculate number of lives saved per year if we were to assume 2000 rates

```python
population_per_year = deaths['est_pop'].unstack().sum()
rate_difference = (sd_rate_3450[2000] - sd_rate_3450) / 100000
lives_saved = rate_difference * population_per_year.ix['2000':'2010']
```

Between 2000 and 2011 the fraction of deaths in San Diego due to these chronic diseases has fallen from 63% to 56% of all deaths on the record.  The percentages seem small, but results of public effors are responsible for __saving over 20,000 lives in a ten year span__.

```python
def timecourse_plot(pct_chronic, lives_saved, title=''):
    fig, axs = subplots(2,1, figsize=(8,5), sharex=True)
    pct_chronic.dropna().plot(kind='bar', ax=axs[0])
    axs[0].set_ybound(.4)
    axs[0].set_ylabel('3-4-50 Death %')
    
    lives_saved.dropna().plot(kind='bar', ax=axs[1], color=colors[1])
    axs[1].set_ylabel('# of Lives Saved')
    for ax in axs:
        prettify_ax(ax)
    axs[0].set_title(title)
```

```python
timecourse_plot(pct_chronic, lives_saved)
```

```python
lives_saved.sum()
```

## Project future life savings given 50% 3-4-50 goal.

* If we reach our goal of 50% we can save __1000 more lives__ per year as well as countless tax dollars. 
* I am using 2010 rates and populations here and comparing to 2000 rates so we can show them on the same plot.  

```python
projected_rate = ((sd_rate_3450[2000] - (sd_rate_all.ix[2010] / 2)) / 100000) 
projected_lives_saved = projected_rate * population_per_year.ix[2010]
projected_lives_saved
```

```python
projected_lives_saved - lives_saved[2010]
```

Update plot with projection data

```python
pct_chronic_proj = pct_chronic.set_value('Goal', .5)
lives_saved_proj = lives_saved.set_value('Goal', projected_lives_saved)
```

```python
timecourse_plot(pct_chronic_proj, lives_saved_proj)
plt.savefig('./figs/San_Diego_County_timecourse.png')
```

## Breaking it down by district

Here I am doing the same operation as before but with individual districts.

```python
district_rate_3450 = (deaths['deaths3450'].unstack().astype(float) / 
                      deaths['est_pop'].unstack()) * 100000
district_rate_all = (deaths['all'].unstack().astype(float) / 
                     deaths['est_pop'].unstack()) * 100000
district_pct_chronic = (district_rate_3450 / district_rate_all).ix[:, '2000':'2010']
```

```python
population_per_year = deaths['est_pop'].unstack()
rate_difference = (district_rate_3450.ix[:,2000] - district_rate_3450.T).T / 100000
distric_lives_saved = rate_difference * population_per_year.ix[:, '2000':'2010']
```

Number of lives saved by district from 2001-2010.  Not adjusted for population across districts so larger district should have high numbers.

```python
def district_plot(vec, ylabel=''):
    fig, ax = subplots(figsize=(12,4))
    vec.dropna().order().plot(kind='bar', ax=ax)
    ax.set_xlabel('')
    ax.set_ylabel(ylabel)
    prettify_ax(ax)
```

```python
district_plot(distric_lives_saved.sum(1), 'Number of Lives Saved');
plt.savefig('./figs/Lives_Saved_by_District.png')
```

Now normalizing to 100,000 residents per city.  Something went right in La Mesa.

```python
district_plot((rate_difference.sum(1)) * 100000, 'Number of Lives Saved (per 100k)');
plt.savefig('./figs/Lives_Saved_by_District_Normalized.png')
```

Year by year plots per district

```python
sra = 'La Mesa'
timecourse_plot(district_pct_chronic.ix[sra], distric_lives_saved.ix[sra])
```

Projected number of lives saved per year if 50% goal is reached.  Not adjusted for population across districts so larger district should have high numbers.

```python
projected_rate_d = ((district_rate_3450.ix[:,2000] - 
                     (district_rate_all.ix[:,2010] / 2)) / 100000) 
projected_lives_saved_d = projected_rate_d * population_per_year.ix[:, 2010]
```

```python
district_plot(projected_lives_saved_d, 'Number of Lives Saved vs. 2000');
```

Now comparing to 2010 death rates.

```python
district_plot((projected_lives_saved_d - distric_lives_saved.ix[:, 2010]), 
              'Number of Lives Saved vs 2010')
plt.savefig('./figs/Number_of_lives_saved_per_year_at_50_goal.png')
```

Update year by year plots with projection data

```python
pct_chronic_proj_d = district_pct_chronic.copy()
pct_chronic_proj_d['Goal'] = .5
lives_saved_proj_d = distric_lives_saved
lives_saved_proj_d['Goal'] = projected_lives_saved_d
```

### Timecourse Plots by District

```python
for sra in pct_chronic_proj_d.dropna().index:
    timecourse_plot(pct_chronic_proj_d.ix[sra], lives_saved_proj_d.ix[sra])
    plt.suptitle(str(sra), size=16)
    plt.savefig('./figs/{}_timecourse.png'.format(sra.replace('/', '_')))
```
